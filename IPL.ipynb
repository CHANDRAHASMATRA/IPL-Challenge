```python
def avgfinder(batsmen,bowler):
    with open('C:\\Users\\MATRA\\OneDrive\\Desktop\\all matches -1.csv', mode='r') as csv_file:
        csv_reader = csv.reader(csv_file)
        line_count = 0
        dummy=0
        ave=0
        for row in csv_reader:
            if line_count == 0:
                pass
                line_count += 1
            if row[8]==batsmen:
                if row[10]==bowler:
                    ave=ave+int(row[11])
                    dummy=dummy+1

            line_count += 1
        return(ave/dummy)
        
```
